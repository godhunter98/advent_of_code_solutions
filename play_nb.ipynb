```python
def move_container(stacks:list,fr:int,to:int,count:int)->list:
    fr -=1
    to -=1
    for i in range(count):
        tbm = stacks[fr].pop()
        stacks[to].append(tbm)
    return stacks

move_container(stacks,1,3,1)
```

```python
steps = '''move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
'''

each_step = steps.splitlines()

stacks = '''
    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 '''
```

```python
def build_stack(container_stack):
    lines = container_stack.strip('\n').splitlines()
    crate_rows = lines[:-1]
    num_stacks = len(lines[-1].split())
    stacks = [[]for _ in range(num_stacks)]

    for row in crate_rows:
        for i in range(num_stacks):
            chunk = row[i*4:(i+1)*4]
            if '[' in chunk:
                crate = chunk[1]
                stacks[i].append(crate)
    for s in stacks:
        s.reverse()
    return stacks
```

```python
useful_stack = build_stack(stacks)
for step in each_step:
    _,count,_,fr,_,to = (step.split(' '))
    move_container(useful_stack,int(fr),int(to),int(count))
    
```
